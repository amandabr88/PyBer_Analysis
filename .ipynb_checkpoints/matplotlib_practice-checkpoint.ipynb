```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
#Set the x-axis to a list of string for each month.
x_axis = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

#Set the y-axis to a list of float as the total fare in the US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Plot a Line Chart
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
# object-oriented interface method
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
# object-oriented interface method
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
# To create a quick graph using the object-oriented interface method
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot using plt.show()
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Change figure size
#plt.figure(figsize=(20, 10))
#plt.subplots(figsize=(20, 10))

# Create the plot and add a label for the legend.
# add a new line color, modified line width, and a marker for our data points
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)
#plt.xlim("Jan","Jun")
#plt.xlim(0, 5)

# Create a title.
plt.title("PyBer Fare by Month")

# Add a grid.
plt.grid()

# Add the legend.
plt.legend()
```

```python
# Change figure size
#plt.figure(figsize=(20, 10))
fig, ax = plt.subplots(figsize=(20, 10))

# Create the plot and add a label for the legend.
# add a new line color, modified line width, and a marker for our data points
ax.plot(x_axis, y_axis, marker = "d", color="green", linewidth=2, label='Boston')

# Create labels for the x and y axes.
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")

# Set the y limit between 0 and 45.
ax.set_ylim(0, 45)
#plt.xlim("Jan","Jun")
#plt.xlim(0, 5)

# Create a title.
ax.set_title("PyBer Fare by Month")

# Add a grid.
ax.grid()

# Add the legend.
ax.legend()
```

```python
# Create a Bubble Chart Using the Object-Oriented Interface - Skill Drill
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i * 5 for i in y_axis], cmap="SkyBlue", edgecolor="k", label='Boston', linewidth=2, alpha=0.2)

#sort Y
ax.invert_yaxis()

# Lables
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Months')
ax.set_ylabel('Fare($)')
ax.set_xlim(0,50)
ax.legend()
```

```python
# Create a Pie Chart Using the MATLAB Method
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Create a Pie Chart Using the MATLAB Method
plt.subplots(figsize=(8, 8))

# Add percentages for each month and "explode" the largest percentage, which is July, the seventh value in the x_axis.
# The "explode" parameter will offset the indicated wedge by a fraction of the radius 
## where "0" is zero distance from the center and "1" is completely outside the diameter of the pie.
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

# Add explode variable and format the number
## autopct = format of one decimal place using .1f%. S
## The % before and after the .1f% formats the number as a percentage.'%.1f%%'
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# Create a Pie Chart Using the Object-Oriented Interface
fig, ax = plt.subplots(figsize=(8, 8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
ax.pie(y_axis,
       labels=x_axis, 
       autopct='%.1f%%',
       explode=explode_values,
       colors=colors,
       shadow=True,
       startangle=90,
       counterclock=False
       
)
plt.show()
```
